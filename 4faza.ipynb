```python
import numpy as np
import pandas as pd

# Трехточечные оценки для каждой задачи
tasks = [
    {"name": "Разработка архитектуры системы", "O": 40, "M": 60, "P": 100},
    {"name": "Настройка балансировщика нагрузки", "O": 20, "M": 32, "P": 48},
    {"name": "Настройка веб-серверов", "O": 24, "M": 40, "P": 64},
    {"name": "Разработка сервиса аутентификации", "O": 48, "M": 80, "P": 120},
    {"name": "Разработка сервиса управления курсами", "O": 60, "M": 100, "P": 160},
    {"name": "Разработка сервиса оценок", "O": 40, "M": 60, "P": 100},
    {"name": "Интеграция с MongoDB", "O": 32, "M": 48, "P": 80},
    {"name": "Настройка Amazon S3", "O": 20, "M": 32, "P": 48},
    {"name": "Настройка HashiCorp Vault", "O": 20, "M": 32, "P": 48},
    {"name": "Настройка NFS", "O": 16, "M": 24, "P": 40},
    {"name": "Разработка интерфейса пользователя", "O": 60, "M": 100, "P": 160},
    {"name": "Тестирование системы", "O": 40, "M": 60, "P": 100},
    {"name": "Написание документации", "O": 20, "M": 32, "P": 48},
    {"name": "Развёртывание в тестовой среде", "O": 20, "M": 32, "P": 48},
    {"name": "Развёртывание в продакшн", "O": 20, "M": 32, "P": 48},
    {"name": "Мониторинг и логирование", "O": 32, "M": 48, "P": 80},
    {"name": "Оптимизация производительности", "O": 40, "M": 60, "P": 100},
    {"name": "Обучение пользователей", "O": 20, "M": 32, "P": 48},
    {"name": "Поддержка и обновление системы", "O": 40, "M": 60, "P": 100},
    {"name": "Анализ обратной связи и доработка", "O": 40, "M": 60, "P": 100},
    {"name": "Интеграция с геосервисами", "O": 48, "M": 80, "P": 120},
    {"name": "Разработка системы уведомлений", "O": 32, "M": 48, "P": 80},
    {"name": "Разработка API для мобильного приложения", "O": 40, "M": 60, "P": 100},
    {"name": "Настройка CI/CD", "O": 32, "M": 48, "P": 80},
    {"name": "Оптимизация базы данных", "O": 40, "M": 60, "P": 100}
]

# Функция для симуляции продолжительности задачи
def simulate_task_duration(O, M, P):
    return np.random.triangular(O, M, P)

# Симуляция выполнения всех задач
def monte_carlo_simulation(tasks, iterations=10000):
    total_durations = []
    for _ in range(iterations):
        total_duration = sum(simulate_task_duration(task["O"], task["M"], task["P"]) for task in tasks)
        total_durations.append(total_duration)
    return total_durations

# Запуск симуляции
durations = monte_carlo_simulation(tasks)

# Оценка с 80% достоверностью
confidence_interval = np.percentile(durations, [10, 90])
print(f"80% Confidence Interval: {confidence_interval[0]:.2f} - {confidence_interval[1]:.2f} часов")

# Среднее значение и стандартное отклонение
mean_duration = np.mean(durations)
std_duration = np.std(durations)
print(f"Среднее время выполнения: {mean_duration:.2f} часов")
print(f"Стандартное отклонение: {std_duration:.2f} часов")

# Построение гистограммы
import matplotlib.pyplot as plt
plt.hist(durations, bins=50, edgecolor='black')
plt.axvline(confidence_interval[0], color='red', linestyle='dashed', linewidth=2, label='10%')
plt.axvline(confidence_interval[1], color='green', linestyle='dashed', linewidth=2, label='90%')
plt.axvline(mean_duration, color='blue', linestyle='dashed', linewidth=2, label='Среднее')
plt.xlabel('Общее время выполнения (часы)')
plt.ylabel('Частота')
plt.legend()
plt.title('Распределение общего времени выполнения задач')
plt.show()
```

```python
# Пример: количество задач, выполненных за неделю
weekly_tasks_completed = [3, 4, 5, 4, 3, 5, 4, 3, 4, 5]

# Симуляция выполнения задач
def simulate_task_completion(weekly_tasks, total_tasks, iterations=10000):
    weeks_to_complete = []
    for _ in range(iterations):
        weeks = 0
        remaining_tasks = total_tasks
        while remaining_tasks > 0:
            tasks_completed = np.random.choice(weekly_tasks)
            remaining_tasks -= tasks_completed
            weeks += 1
        weeks_to_complete.append(weeks)
    return weeks_to_complete

# Запуск симуляции
weeks_to_complete = simulate_task_completion(weekly_tasks_completed, len(tasks))

# Оценка с 80% достоверностью
weeks_confidence_interval = np.percentile(weeks_to_complete, [10, 90])
print(f"80% Confidence Interval: {weeks_confidence_interval[0]:.2f} - {weeks_confidence_interval[1]:.2f} недель")
```

```python
# Пример: количество человеко-часов в неделю
weekly_hours = [120, 140, 160, 180, 200]

# Симуляция выполнения задач с учётом трудозатрат
def simulate_hours_completion(weekly_hours, tasks, iterations=10000):
    weeks_to_complete = []
    for _ in range(iterations):
        total_hours = sum(simulate_task_duration(task["O"], task["M"], task["P"]) for task in tasks)
        weeks = 0
        remaining_hours = total_hours
        while remaining_hours > 0:
            hours_completed = np.random.choice(weekly_hours)
            remaining_hours -= hours_completed
            weeks += 1
        weeks_to_complete.append(weeks)
    return weeks_to_complete

# Запуск симуляции
weeks_to_complete_hours = simulate_hours_completion(weekly_hours, tasks)

# Оценка с 80% достоверностью
weeks_confidence_interval_hours = np.percentile(weeks_to_complete_hours, [10, 90])
print(f"80% Confidence Interval: {weeks_confidence_interval_hours[0]:.2f} - {weeks_confidence_interval_hours[1]:.2f} недель")
```
